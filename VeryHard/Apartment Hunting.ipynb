    **TLDR:** 
    Find the distance between city blocks from the apartment to the amenities,\
    find the apartment closest to any given amenity.

*Full prompt below code.



```python
def apartmentHunting(blocks, reqs):
    # create a list of dictionaries as a copy of 'blocks' 
    # to later set the required distances to each amenity
    reqDs = [block for block in blocks]
    # print(reqDs)

    # iterate FORWARDS through list of blocks
    for block in range(len(blocks)):
      # and through each amenity
        for req in reqs:
          # if the required amenity is on the block
            if blocks[block][req] == True:
                #set required walking distance to zero,
                reqDs[block][req] = 0 

            elif blocks[block][req] == False:
                # if required amenity is not on the first block
                # set distance to infinity.
                if block == 0:
                    reqDs[block][req] = float('inf')
                #if the amenity is not on subsequent blocks, set it to one more than previous
                else:
                    reqDs[block][req] = reqDs[block - 1][req] + 1
                
    # iterate BACKWARDS through list of blocks       
    for block in range(len(blocks) -1, 0, -1):
        #and through each amenity
        for req in reqs:
            # compare distances between amenities, if the current block is closer to the 
            # given amenity make sure it is no more than + 1 compared to current distance
            # ex. ['inf', 0, 1, 2, 0]
            #     [ a   , b, c, d, e] when comparing slot 'e' to 'd', d should only be 1 instead of 2
            if reqDs[block - 1][req] > reqDs[block][req]:
                reqDs[block - 1][req] = reqDs[block][req] + 1

    maxPerBlock = [ float('-inf') for _ in reqDs]
    
    for block in range(len(reqDs)):
        #make required Distances into a list of lists, rather than dictionaries.
        distanceToAmenity = []
        for req in reqDs[block]:
            distanceToAmenity.append(reqDs[block][req])
        # print(distanceToAmenity)

        # set the max per block to the greatest distance of any amenity
        maxPerBlock[block] = max(distanceToAmenity)

    #this print helps visualize which block has shortest distance to any given amenity.
    # print(maxPerBlock)    

    minDistance = min(maxPerBlock)
    for idx in range(len(maxPerBlock)):
        if maxPerBlock[idx] == minDistance:
            # Print the conclusion:
            print("The most centralized block is block", idx, "to reach any and all required amenities.")
            #Return ideal block idx
            return idx

# def pprint(dictList):
#     for dict in dictList:
#         print(dict)

blocks = [
  { "gym": False,
    "school": True,
    "store": False   },

  { "gym": True,
    "school": False,
    "store": False  },

  { "gym": True,
    "school": True,
    "store": False },

  { "gym": False,
    "school": True,
    "store": False },

  { "gym": False,
    "school": True,
    "store": True   }
]
reqs = ["gym", "school", "store"]
apartmentHunting(blocks, reqs)
```
